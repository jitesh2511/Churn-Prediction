# Data Exploration Summary

### Columns Description
- **customerID**: 7043 unique values  
  - ID of the customer
- **gender**: Male, Female  
  - Gender of the customer
- **SeniorCitizen**: 1, 0  
  - Senior citizen status (1 = Yes, 0 = No)
- **Partner**: Yes, No  
  - Has a partner or spouse
- **Dependents**: Yes, No  
  - Has dependents (children, elderly parents)
- **tenure**: Numeric  
  - Months the customer has stayed
- **PhoneService**: Yes, No  
  - Has phone service
- **MultipleLines**: Yes, No, No phone service  
  - Multiple phone lines
- **InternetService**: Fiber Optic, DSL, No  
  - Internet provider
- **OnlineSecurity**: Yes, No, No internet service  
  - Has online security
- **OnlineBackup**: Yes, No, No internet service  
  - Has online backup
- **DeviceProtection**: Yes, No, No internet service  
  - Has device protection
- **TechSupport**: Yes, No, No internet service  
  - Has tech support
- **StreamingTV**: Yes, No, No internet service  
  - Has streaming TV
- **StreamingMovies**: Yes, No, No internet service  
  - Has streaming movies
- **Contract**: Month-to-month, One year, Two year  
  - Type of contract
- **PaperlessBilling**: Yes, No  
  - Uses paperless billing
- **PaymentMethod**: Electronic check, Mailed check, Bank transfer (auto), Credit card (auto)  
  - Payment method
- **MonthlyCharges**: Numeric  
  - Amount charged monthly
- **TotalCharges**: Numeric  
  - Total amount charged
- **Churn**: Yes, No  
  - Whether the customer churned


---

### Data Insights

- **Shape**: (7043, 21)
- **Target Variable**: Churn (Binary classification problem)
- **Missing Values**: 11 missing values in `TotalCharges` correspond to customers with zero tenure (MonthlyCharges x 0 = 0)
- **Duplicates**: No duplicated records
- **Numerical Features**: tenure, MonthlyCharges, TotalCharges
- **Categorical Features**: Mix of binary and multi-class categorical variables
- **Binary Features**: Several Yes/No features that can be encoded during preprocessing
- **customerID**: Unique identifier, not useful for prediction (to be dropped)

---

### Important Observations

- **Class Imbalance**: ~27% churn vs ~73% non-churn
